```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
benchmark = pd.read_csv("benchmark_results.csv")
```

```python
benchmark
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,4))
ax[0].plot(benchmark["input"], benchmark["mean_time_us"]*1e-6)
ax[0].set_title("Execution Time for Quadrangulation Sampling")
ax[0].set_xlabel("Number of Faces")
ax[0].set_xscale("log")
ax[0].set_ylabel("Execution Time (s)")
ax[0].set_yscale("log")
ax[1].plot(benchmark["input"], benchmark["mean_peak_mem_kb"]*1e-6)
ax[1].set_title("Peak Allocated Memory for Quadrangulation Sampling")
ax[1].set_xlabel("Number of Faces")
ax[1].set_xscale("log")
ax[1].set_ylabel("Peak Allocated Memory (GB)")
ax[1].set_yscale("log")
plt.tight_layout()
plt.savefig("quadrangulation_sampling.pdf")
plt.show()
```
